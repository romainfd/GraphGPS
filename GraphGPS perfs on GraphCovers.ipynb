```python
from os import listdir
from os.path import isdir, join
import pandas as pd
import json 

dfs = []

dirs = [f for f in listdir('results/graphcovers-history/') if isdir(join('results/graphcovers-history/', f))]
for dirName in dirs:
    d, n = dirName.split('-')
    degree = int(d[1:])
    nb_covers = int(n[1:])
    print(degree, nb_covers)

    with open(f'results/graphcovers-history/{dirName}/test/stats.json') as f:
        data = pd.DataFrame.from_records([json.loads(l) for l in f.readlines()])
        data['degree'] = degree
        data['nb_covers'] = nb_covers
        with open(f'results/graphcovers-history/{dirName}/logging.log') as log:
            data['nb_covers_real'] = int(log.readlines()[6].split("num graphs: ")[1]) // 3
        dfs.append(data)
        
df = pd.concat(dfs)
print(df)
```

```python
max_acc = df.groupby(['degree', 'nb_covers']).max('accuracy')
min_loss = df.groupby(['degree', 'nb_covers']).min('loss')
max_acc
```
